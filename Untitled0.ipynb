b) Кодирование категориальных признаков:

c) Нормализация числовых данных:

Min-Max Scaling (приведение значений к диапазону [0,1]).

Standardization (преобразование к среднему 0 и стандартному отклонению 1).


```python
from sklearn.preprocessing import LabelEncoder

# 1. Проверка и обработка пропущенных значений
missing = df.isnull().sum()
print("Пропущенные значения:\n", missing[missing > 0])

# Заполнение пропусков
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)  # для строк — мода
        else:
            df[col].fillna(df[col].mean(), inplace=True)  # для чисел — среднее

# 2. Кодирование категориальных признаков
# Сначала получим список категориальных колонок
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("Категориальные признаки:", categorical_columns)

# Укажи вручную, какие колонки считаешь упорядоченными (если есть)
ordered_columns = []  # Пример: ["some_ordered_column"]

# Разделим на упорядоченные и неупорядоченные
onehot_columns = [col for col in categorical_columns if col not in ordered_columns]
label_columns = ordered_columns  # пока пустой список, но можно вручную добавить

# Label Encoding — если есть упорядоченные признаки
label_encoder = LabelEncoder()
for col in label_columns:
    df[col] = label_encoder.fit_transform(df[col])

# One-Hot Encoding — для остальных (например, protocol_type, service, flag)
df = pd.get_dummies(df, columns=onehot_columns, drop_first=True)
```

```python
from imblearn.over_sampling import RandomOverSampler # создает новые образцы, выбирая случайные примеры из меньшинственного класса с возвращением
from imblearn.under_sampling import RandomUnderSampler # cлучайным образом выбирает и удаляет примеры из большинства класса, чтобы сбалансировать количество примеров между классами
# RandomOverSampler и RandomUnderSampler — это классы из библиотеки imblearn, которые используются для обработки несбалансированных наборов данных.
# Несбалансированные данные возникают, когда количество примеров в разных классах сильно различается, что может негативно сказаться на производительности моделей машинного обучения.
from collections import Counter # позволяет подсчитывать количество экземпляров каждого класса в наборе данных.
X = data.iloc[:, :-1]  #  все столбцы, кроме последнего используем их в качестве признаков
y = data.iloc[:, -1]   # последний столбец используется как целевая переменная, которую мы хотим предсказать.

# Проверка распределения классов до балансировки
print("Исходное распределение (до балансировки):", Counter(y))

# Oversampling
oversampler = RandomOverSampler(sampling_strategy='auto') # автоматически балансирует классы до одинакового количества.
X_resampled, y_resampled = oversampler.fit_resample(X, y) # создает новые выборки, увеличивая количество экземпляров меньшинства.

print("После Oversampling:", Counter(y_resampled))

# Undersampling
undersampler = RandomUnderSampler(sampling_strategy='auto') # RandomUnderSampler используется для уменьшения количества экземпляров большинства в целевой переменной.
X_resampled, y_resampled = undersampler.fit_resample(X, y) # создает новые выборки, уменьшая количество экземпляров большинства.

print("После Undersampling:", Counter(y_resampled))
```

В машинном обучении часто встречаются несбалансированные данные — когда один класс значительно преобладает над другим (например, 90% одного класса и 10% другого). Это может ухудшить качество модели, так как она будет склоняться к предсказанию более частого класса.

Для решения этой проблемы используют два основных метода:

Oversampling (передискретизация)

Увеличивает количество примеров миноритарного (меньшего) класса, дублируя существующие или создавая синтетические данные (например, SMOTE).

Плюсы: не теряет информацию.

Минусы: может привести к переобучению, если просто копировать данные.

Undersampling (недодискретизация)

Уменьшает количество примеров мажоритарного (большего) класса, случайно удаляя часть данных.

Плюсы: быстрее обучается, так как данных меньше.

Минусы: может потерять важные закономерности из-за удаления данных.
